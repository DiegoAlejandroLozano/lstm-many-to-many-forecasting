# Análisis del Dataset
Se realiza un análisis del dataset con el fin de identificar sus principales características y comprender su estructura general.

## Importación de librerías

```python
import pandas as pd
import numpy  as np
```

## Lectura de los datos

Se realiza la lectura del dataset y una primera inspección visual para obtener una visión general de su contenido y estructura.

```python
datos_df = pd.read_csv("../data/01_datos_raw/LD2011_2014.txt", sep=";")
datos_df
```

Se muestra la información del dataset y se identifica que la mayoría de las columnas aparecen como tipo object, debido a que los valores numéricos utilizan la coma (,) como separador decimal.

```python
datos_df.info(verbose=True)
```

También se verifica que no existen valores nulos en ninguna de las columnas del dataset.

```python
datos_df.isna().sum()
```

# Procesamiento

Se aplican las siguientes transformaciones al dataset:

* La primera columna se renombra como Fecha.

* Se reemplaza la coma (,) por el punto (.) como separador decimal para convertir correctamente los valores numéricos.

* Convertir la columna Fecha a datetime.

* Se crean las columnas minuto, hora, día, mes y año a partir de la columna de fecha.

* Transformación de variables cíclicas.

## Renombrar la primera columna.

```python
datos_df.rename(columns={"Unnamed: 0":"Fecha"}, inplace=True)
datos_df
```

## Reemplazar la coma (,) por el punto (.)

```python
# 1. Eliminar espacios, caracteres invisibles y limpiar string
datos_df = datos_df.apply(lambda col: col.astype(str).str.strip())

# 2. Reemplazar comas por puntos
datos_df.replace(",", ".", regex=True, inplace=True)

# 3. Convertir todas las columnas menos la primera a numéricas
for col in datos_df.columns[1:]:
    datos_df[col] = pd.to_numeric(datos_df[col], errors="coerce")

datos_df.tail(5)
```

Se verifica que las columnas correspondientes a los usuarios cambien su tipo de dato de object a float.

```python
datos_df.info(verbose=True)
```

Se verifica que, después de las transformaciones realizadas, no se generen valores nulos en el dataset.

```python
datos_df.isna().sum()
```

## Convertir la columna Fecha a datetime

```python
datos_df["Fecha"] = pd.to_datetime(datos_df["Fecha"])
datos_df.head(5)
```

## Crear las columnas minuto, hora, día, mes y año a partir de la columna de fecha.

```python
nuevas_cols = pd.DataFrame({
    "minuto"     : datos_df["Fecha"].dt.minute,
    "hora"       : datos_df["Fecha"].dt.hour,
    "dia_año"    : datos_df["Fecha"].dt.dayofyear,
    "dia_semana" : datos_df["Fecha"].dt.dayofweek,
    "mes"        : datos_df["Fecha"].dt.month,
    "año"        : datos_df["Fecha"].dt.year
})

datos_df = pd.concat([datos_df, nuevas_cols], axis=1)
datos_df.head(5)
```

## Transformación de variables cíclicas

Para las variables de naturaleza cíclica, se aplica una transformación mediante funciones seno y coseno, con el fin de representar adecuadamente su periodicidad y evitar discontinuidades en los límites del ciclo.

```python
datos_df["minutos_sin"] = np.sin(2*np.pi*datos_df["minuto"] / 60)
datos_df["minutos_cos"] = np.cos(2*np.pi*datos_df["minuto"] / 60)

datos_df["hora_sin"] = np.sin(2*np.pi*datos_df["hora"] / 24)
datos_df["hora_cos"] = np.cos(2*np.pi*datos_df["hora"] / 24)

datos_df["dia_año_sin"] = np.sin(2*np.pi*datos_df["dia_año"] / 365)
datos_df["dia_año_cos"] = np.cos(2*np.pi*datos_df["dia_año"] / 365)

datos_df["dia_semana_sin"] = np.sin(2*np.pi*datos_df["dia_semana"] / 7)
datos_df["dia_semana_cos"] = np.cos(2*np.pi*datos_df["dia_semana"] / 7)

datos_df["mes_sin"] = np.sin(2*np.pi*datos_df["mes"] / 12)
datos_df["mes_cos"] = np.cos(2*np.pi*datos_df["mes"] / 12)

# Eliminar columnas que no se utilizarán
datos_df.drop(columns=["minuto", "hora", "dia_año", "dia_semana", "mes"], inplace=True)

# Reordenar las columnas
cols_temporales = [
    "Fecha",
    "minutos_sin",
    "minutos_cos",
    "hora_sin",
    "hora_cos",
    "dia_año_sin",
    "dia_año_cos",
    "dia_semana_sin",
    "dia_semana_cos",
    "mes_sin",
    "mes_cos"
]
nuevo_orden = (
    cols_temporales +
    [col for col in datos_df.columns if col not in cols_temporales]
)
datos_df = datos_df[nuevo_orden]

datos_df
```

Se guarda el nuevo dataset

```python
datos_df.to_csv("../data/02_datos_procesados/datos.csv", sep=";", index=False)
```
